```python
import pandas as pd
df = pd.read_csv("/content/Titanic-Dataset.csv")
print("DATA LOADED SUCCESSFULLY\n")

print("INFO OF DATAFRAME:")
print(df.info())
print("\nHEAD:")
print(df.head())

# 3.1 WARM UP EXERCISES

# PROBLEM 1 – SORTING

# 1. Fare DataFrame
fare = df[["Fare"]]
print("\nFare head:")
print(fare.head())

# 2. Pclass + Age DataFrame
class_age = df[["Pclass", "Age"]]
print("\nPclass + Age head:")
print(class_age.head())

# 3. Survived + Sex DataFrame
survived_gender = df[["Survived", "Sex"]]
print("\nSurvived + Sex head:")
print(survived_gender.head())

# PROBLEM 2 – SUBSETTING

# Subsetting rows:
fare_gt_100 = df[df["Fare"] > 100]
print("\nFare > 100:")
print(fare_gt_100.head())

first_class = df[df["Pclass"] == 1]
print("\nFirst class passengers:")
print(first_class.head())

female_under_18 = df[(df["Age"] < 18) & (df["Sex"] == "female")]
print("\nFemales under 18:")
print(female_under_18.head())

# Subsetting by categorical:
embarked_c_or_s = df[df["Embarked"].isin(["C", "S"])]
print("\nEmbarked C or S:")
print(embarked_c_or_s.head())

first_second_class = df[df["Pclass"].isin([1, 2])]
print("\nFirst or Second class:")
print(first_second_class.head())


# 3.2 EXPLORATORY DATA ANALYSIS – PRACTICE EXERCISE 1

# Fill missing Age values with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Question 1: Highest fare relative to age

df["fare_per_year"] = df["Fare"] / df["Age"]

high_fare_age = df[df["fare_per_year"] > 5]

high_fare_age_srt = high_fare_age.sort_values(by="fare_per_year", ascending=False)

result1 = high_fare_age_srt[["Name", "fare_per_year"]]

print("\nHighest fare paid relative to age:")
print(result1.head())

# Question 2: Highest fare relative to class (adult men)

df["fare_per_class"] = df["Fare"] / df["Pclass"]

adult_males = df[(df["Sex"] == "male") & (df["Age"] >= 18)]

adult_males_srt = adult_males.sort_values(by="fare_per_class", ascending=False)

result2 = adult_males_srt[["Name", "Age", "fare_per_class"]]

print("\nAdult male with highest fare relative to class:")
print(result2.head())

# 3.3 GROUP BY PRACTICE EXERCISE

# Question 1: Percent of total fare revenue from each class

total_fare = df["Fare"].sum()

fare_class1 = df[df["Pclass"] == 1]["Fare"].sum()
fare_class2 = df[df["Pclass"] == 2]["Fare"].sum()
fare_class3 = df[df["Pclass"] == 3]["Fare"].sum()

fare_list = [fare_class1, fare_class2, fare_class3]

proportions = [(value / total_fare) * 100 for value in fare_list]

print("\nPercentage of fare revenue by class:")
print(f"Class 1: {proportions[0]:.2f}%")
print(f"Class 2: {proportions[1]:.2f}%")
print(f"Class 3: {proportions[2]:.2f}%")

# Question 2: Percent of passengers by age group

def age_group(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

df["age_group"] = df["Age"].apply(age_group)

total_passengers = len(df)

age_group_counts = df["age_group"].value_counts()

age_group_percent = (age_group_counts / total_passengers) * 100

print("\nPercentage of passengers by age group:")
print(age_group_percent)
```
